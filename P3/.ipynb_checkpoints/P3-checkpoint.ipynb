As discussed the data set consists of the following timeseries:
- water_temp (the response)
- measurements of 7 atmospheric values... the naming convention is "station_id"(where the measurement is done) + "value"(what got measured) + "number"(0=measurement in 2m height, 1=measurement in 10 height above ground)

We have
- water_temp in degC - measured at 7,10,14, 16 every day during swimming season

- air_temperature in degC
- wind_speed in m/s
- cloud_area_fraction in 0,...,8 (0 no clouds, 8 fully cloudy)
- mean(solar irradiance) in w/m2 - a quantity for the intensity of the sun
- sum(duration of sunshine) in min - sunshine in the last hour
- mean(realtive humidity) in %
- mean(downwelling shortwave flux) in W/m2 - like irradiance but only for UV radiation (normally criterion for getting sun burned, but I thought that it could maybe get interesting)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Make numpy printouts easier to read.
np.set_printoptions(precision=3, suppress=True)

import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing

print(tf.__version__)
```

```python
# Import data
column_names = ['water_temp', 'air_temp1', 'air_temp2', 'air_temp3', 'air_temp4',
                'wind_speed1', 'wind_speed2', 'wind_speed3', 'cloud1', 'cloud2',
                'cloud3', 'cloud4', 'solar_mean', 'solar_sum', 'humidity1', 'humidity2',
                'surface_downwelling']

raw_dataset = pd.read_csv('dataset1.csv', names = column_names)
raw_dataset = raw_dataset.iloc[1:]
dataset = raw_dataset.copy()
# dataset.tail()
dataset.isna().sum()
# dataset = dataset.dropna()
# df["A"] = pd.to_numeric(df["A"], downcast="float")
del dataset["wind_speed3"]
del dataset["cloud2"]

dataset.tail()
for i in range(len(dataset.columns)):
    dataset.iloc[:,i] = pd.to_numeric(dataset.iloc[:,i], downcast = "float")
dataset
```

```python
# Split the data set
train_dataset = dataset.sample(frac=0.8, random_state=0)
test_dataset = dataset.drop(train_dataset.index)
train_dataset.tail()
```

```python
sns.pairplot(train_dataset, vars = train_dataset.columns[:], diag_kind='kde')
plt.show()
```

```python
train_dataset.describe().transpose()

train_features = train_dataset.copy()
test_features = test_dataset.copy()

train_labels = train_features.pop('water_temp')
test_labels = test_features.pop('water_temp')

train_dataset.describe().transpose()
```

```python
normalizer = preprocessing.Normalization()
normalizer.adapt(np.array(train_features))

first = np.array(train_features[:1])
with np.printoptions(precision=2, suppress=True):
  print('First example:', first)
  print()
  print('Normalized:', normalizer(first).numpy())
```

```python
# start with one variable
air_temp1 = train_features['air_temp1']
air_temp1_normalizer = preprocessing.Normalization(input_shape=[1, ])
air_temp1_normalizer.adapt(air_temp1)

air_temp1_model = tf.keras.Sequential([air_temp1_normalizer, layers.Dense(units = 1)])
air_temp1_model.summary()
air_temp1_model.predict(air_temp1[:10])

air_temp1_model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.1), loss = "mean_absolute_error")
history = air_temp1_model.fit(np.array(train_features['air_temp1']).astype(np.float), train_labels,
                              epochs = 100, verbose = 0, validation_split=0.2)#%%
```

```python
# visualise the model's training progress
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()

def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.ylim([0, 10])
  plt.xlabel('Epoch')
  plt.ylabel('Error [MPG]')
  plt.legend()
  plt.grid(True)

plot_loss(history)
```

```python
test_results = {}

test_results['air_temp1_model'] = air_temp1_model.evaluate(
    test_features['air_temp1'],
    test_labels, verbose=0)

x = tf.linspace(0.0, 250, 251)
y = air_temp1_model.predict(x)

def plot_air_temp1(x, y):
  plt.scatter(train_features['air_temp1'], train_labels, label='Data')
  plt.plot(x, y, color='k', label='Predictions')
  plt.xlabel('Air_temp1')
  plt.ylabel('water_temp')
  plt.legend()

plot_air_temp1(x,y)
```

```python
linear_model = tf.keras.Sequential([
    normalizer,
    layers.Dense(units=1)
])

linear_model.predict(train_features[:10])
linear_model.layers[1].kernel
linear_model.compile(
    optimizer=tf.optimizers.Adam(learning_rate=0.1),
    loss='mean_absolute_error')

history = linear_model.fit(
    train_features, train_labels,
    epochs=100,
    # suppress logging
    verbose=0,
    # Calculate validation results on 20% of the training data
    validation_split = 0.2)

plot_loss(history)
```

```python
test_results['linear_model'] = linear_model.evaluate(
    test_features, test_labels, verbose=0)

print(test_results['linear_model'])
```
